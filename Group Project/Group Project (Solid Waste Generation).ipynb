1. Polulation Predection of Calgary for 2024 using Linear regression

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Read data from Excel file
df = pd.read_excel('/content/population of calgary.xlsx', engine='openpyxl')
print(df.columns)
df.columns = df.columns.str.strip()

# Prepare features (X) and target variable (y)
df = df.dropna(subset=['Population'])
X = df["Year"]  # Features: Year
y = df["Population"]  # Target: Population

# Ensure X is a 2D array and y is a 1D array
X = X.values.reshape(-1, 1)
y = y.values

# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predict population for 2024
year_2024 = np.array([[2024]])  # Feature for prediction
population_2024 = model.predict(year_2024)

# Evaluate the model
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Display results
print("Model Performance:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R² Score: {r2:.2f}")
print("\nPredicted Population for 2024:")
print(f"{population_2024[0]:,.0f}")

# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color="blue", label="Actual Population")
plt.plot(X, y_pred, color="red", label="Regression Line")
plt.scatter(year_2024, population_2024, color="green", label="Prediction for 2024", s=100)
plt.title("Calgary Population Prediction")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.grid()
plt.show()
```

2. Waste Predection of Calgary using Linear Regression Model for 2024

```python
# Read data from CSV file
csv_file = '/content/Annual_Waste_Volumes_20241104.csv'
df = pd.read_csv(csv_file)

# Ensure the CSV has the following columns: "Year", "Black Cart (tonnes)", "Blue Cart (tonnes)", "Green Cart (tonnes)"
df = df.dropna(subset=['Year'])
# Prepare features (X) and targets (y)
X = df[["Year"]]  # Features: Year
y_black = df["Black Cart (tonnes)"]  # Target for Black Cart
y_blue = df["Blue Cart (tonnes)"]   # Target for Blue Cart
y_green = df["Green Cart (tonnes)"] # Target for Green Cart

# Initialize and train separate Linear Regression models for each waste cart
model_black = LinearRegression()
model_blue = LinearRegression()
model_green = LinearRegression()

model_black.fit(X, y_black)
model_blue.fit(X, y_blue)
model_green.fit(X, y_green)

# Predict waste for 2024
year_2024 = np.array([[2024]])  # Feature for prediction
black_cart_2024 = model_black.predict(year_2024)
blue_cart_2024 = model_blue.predict(year_2024)
green_cart_2024 = model_green.predict(year_2024)

# Predict waste for 2024
year_2024 = np.array([[2024]])  # Feature for prediction
black_cart_2024 = model_black.predict(year_2024)
blue_cart_2024 = model_blue.predict(year_2024)
green_cart_2024 = model_green.predict(year_2024)

# Evaluate the models
y_pred_black = model_black.predict(X)
y_pred_blue = model_blue.predict(X)
y_pred_green = model_green.predict(X)

mse_black = mean_squared_error(y_black, y_pred_black)
mse_blue = mean_squared_error(y_blue, y_pred_blue)
mse_green = mean_squared_error(y_green, y_pred_green)

r2_black = r2_score(y_black, y_pred_black)
r2_blue = r2_score(y_blue, y_pred_blue)
r2_green = r2_score(y_green, y_pred_green)

# Display results
print("Model Performance:")
print(f"Black Cart - MSE: {mse_black:.2f}, R²: {r2_black:.2f}")
print(f"Blue Cart  - MSE: {mse_blue:.2f}, R²: {r2_blue:.2f}")
print(f"Green Cart - MSE: {mse_green:.2f}, R²: {r2_green:.2f}")

print("\nPredicted Waste for 2024 (tonnes):")
print(f"Black Cart: {black_cart_2024[0]:,.0f}")
print(f"Blue Cart: {blue_cart_2024[0]:,.0f}")
print(f"Green Cart: {green_cart_2024[0]:,.0f}")

# Visualization
plt.figure(figsize=(12, 8))

# Black Cart
plt.subplot(3, 1, 1)
plt.scatter(X, y_black, color="blue", label="Actual Black Cart")
plt.plot(X, y_pred_black, color="red", label="Regression Line")
plt.scatter(year_2024, black_cart_2024, color="green", label="Prediction for 2024", s=100)
plt.title("Black Cart Waste Prediction")
plt.xlabel("Year")
plt.ylabel("Waste (tonnes)")
plt.legend()

# Blue Cart
plt.subplot(3, 1, 2)
plt.scatter(X, y_blue, color="blue", label="Actual Blue Cart")
plt.plot(X, y_pred_blue, color="red", label="Regression Line")
plt.scatter(year_2024, blue_cart_2024, color="green", label="Prediction for 2024", s=100)
plt.title("Blue Cart Waste Prediction")
plt.xlabel("Year")
plt.ylabel("Waste (tonnes)")
plt.legend()

# Green Cart
plt.subplot(3, 1, 3)
plt.scatter(X, y_green, color="blue", label="Actual Green Cart")
plt.plot(X, y_pred_green, color="red", label="Regression Line")
plt.scatter(year_2024, green_cart_2024, color="green", label="Prediction for 2024", s=100)
plt.title("Green Cart Waste Prediction")
plt.xlabel("Year")
plt.ylabel("Waste (tonnes)")
plt.legend()

plt.tight_layout()
plt.show()

```

3. Waste Predection of Calgary using Random Forest Model for 2024

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load data from CSV file
# Replace 'waste_data.csv' with the path to your CSV file
csv_file = "/content/Annual_Waste_Volumes_20241104.csv"
df = pd.read_csv(csv_file)

# Check the first few rows to verify the data
print("Data from CSV:")
print(df.head())

# Ensure the CSV has columns named 'Year', 'Black Cart (tonnes)', 'Blue Cart (tonnes)', and 'Green Cart (tonnes)'
# If not, adjust the column names in the code below accordingly
df = df.dropna(subset=['Black Cart (tonnes)', 'Blue Cart (tonnes)', 'Green Cart (tonnes)'])
# Prepare features (X) and targets (y)
X = df[["Year"]]  # Features: Year
y_black = df["Black Cart (tonnes)"]  # Target for Black Cart
y_blue = df["Blue Cart (tonnes)"]   # Target for Blue Cart
y_green = df["Green Cart (tonnes)"] # Target for Green Cart

# Initialize and train separate Random Forest models for each waste cart
model_black_rf = RandomForestRegressor(n_estimators=100, random_state=42, max_depth=5)
model_blue_rf = RandomForestRegressor(n_estimators=100, random_state=42, max_depth=5)
model_green_rf = RandomForestRegressor(n_estimators=100, random_state=42, max_depth=5)

model_black_rf.fit(X, y_black)
model_blue_rf.fit(X, y_blue)
model_green_rf.fit(X, y_green)

# Predict waste for 2024
year_2024 = np.array([[2024]])  # Feature for prediction
black_cart_2024_rf = model_black_rf.predict(year_2024)
blue_cart_2024_rf = model_blue_rf.predict(year_2024)
green_cart_2024_rf = model_green_rf.predict(year_2024)

# Evaluate the models
y_pred_black_rf = model_black_rf.predict(X)
y_pred_blue_rf = model_blue_rf.predict(X)
y_pred_green_rf = model_green_rf.predict(X)

mse_black_rf = mean_squared_error(y_black, y_pred_black_rf)
mse_blue_rf = mean_squared_error(y_blue, y_pred_blue_rf)
mse_green_rf = mean_squared_error(y_green, y_pred_green_rf)
r2_black_rf = r2_score(y_black, y_pred_black_rf)
r2_blue_rf = r2_score(y_blue, y_pred_blue_rf)
r2_green_rf = r2_score(y_green, y_pred_green_rf)

# Display results
print("Model Performance (Random Forest):")
print(f"Black Cart - MSE: {mse_black_rf:.2f}, R²: {r2_black_rf:.2f}")
print(f"Blue Cart  - MSE: {mse_blue_rf:.2f}, R²: {r2_blue_rf:.2f}")
print(f"Green Cart - MSE: {mse_green_rf:.2f}, R²: {r2_green_rf:.2f}")
print("\nPredicted Waste for 2024 (tonnes):")
print(f"Black Cart: {black_cart_2024_rf[0]:,.0f}")
print(f"Blue Cart: {blue_cart_2024_rf[0]:,.0f}")
print(f"Green Cart: {green_cart_2024_rf[0]:,.0f}")


# Visualization
plt.figure(figsize=(12, 8))

# Black Cart
plt.subplot(3, 1, 1)
plt.scatter(X, y_black, color="blue", label="Actual Black Cart")
plt.plot(X, y_pred_black_rf, color="red", label="Random Forest Prediction")
plt.scatter(year_2024, black_cart_2024_rf, color="green", label="Prediction for 2024", s=100)
plt.title("Black Cart Waste Prediction - Random Forest")
plt.xlabel("Year")
plt.ylabel("Waste (tonnes)")
plt.legend()
plt.savefig("black_cart_rf.png")

# Blue Cart
plt.subplot(3, 1, 2)
plt.scatter(X, y_blue, color="blue", label="Actual Blue Cart")
plt.plot(X, y_pred_blue_rf, color="red", label="Random Forest Prediction")
plt.scatter(year_2024, blue_cart_2024_rf, color="green", label="Prediction for 2024", s=100)
plt.title("Blue Cart Waste Prediction - Random Forest")
plt.xlabel("Year")
plt.ylabel("Waste (tonnes)")
plt.legend()
plt.savefig("blue_cart_rf.png")

# Green Cart
plt.subplot(3, 1, 3)
plt.scatter(X, y_green, color="blue", label="Actual Green Cart")
plt.plot(X, y_pred_green_rf, color="red", label="Random Forest Prediction")
plt.scatter(year_2024, green_cart_2024_rf, color="green", label="Prediction for 2024", s=100)
plt.title("Green Cart Waste Prediction - Random Forest")
plt.xlabel("Year")
plt.ylabel("Waste (tonnes)")
plt.legend()
plt.savefig("green_cart_rf.png")

plt.tight_layout()
plt.show()
```
